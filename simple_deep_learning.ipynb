```python
import numpy as np

class Node:
    def __init__(self):
        self.value = None
        self.grad = None

    def forward(self):
        pass

    def backward(self):
        pass


class Linear(Node):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.W = np.random.randn(output_size, input_size) * 0.1
        self.b = np.zeros((output_size, 1))
        self.x = None
        self.dW = None
        self.db = None

    def forward(self, x):
        self.x = x
        self.value = np.dot(self.W, x) + self.b
        return self.value

    def backward(self, grad_output):
        self.dW = np.dot(grad_output, self.x.T)
        self.db = grad_output
        grad_input = np.dot(self.W.T, grad_output)
        return grad_input

    def update(self, lr):
        self.W -= lr * self.dW
        self.b -= lr * self.db


class Sigmoid(Node):
    def forward(self, x):
        self.value = 1 / (1 + np.exp(-x))
        return self.value

    def backward(self, grad_output):
        return grad_output * self.value * (1 - self.value)


class ReLU(Node):
    def forward(self, x):
        self.input = x
        self.value = np.maximum(0, x)
        return self.value

    def backward(self, grad_output):
        return grad_output * (self.input > 0)


class MSELoss(Node):
    def forward(self, prediction, target):
        self.prediction = prediction
        self.target = target
        self.value = 0.5 * np.sum((prediction - target) ** 2)
        return self.value

    def backward(self):
        return self.prediction - self.target


class SimpleNeuralNetwork:
    def __init__(self):
        self.layer1 = Linear(2, 3)
        self.activation1 = Sigmoid()
        self.layer2 = Linear(3, 1)
        self.activation2 = Sigmoid()
        self.loss = MSELoss()

    def forward(self, x):
        z1 = self.layer1.forward(x)
        a1 = self.activation1.forward(z1)
        z2 = self.layer2.forward(a1)
        output = self.activation2.forward(z2)
        return output

    def compute_error(self, x, y):
        prediction = self.forward(x)
        return self.loss.forward(prediction, y)

    def backward(self):
        grad = self.loss.backward()
        grad = self.activation2.backward(grad)
        grad = self.layer2.backward(grad)
        grad = self.activation1.backward(grad)
        self.layer1.backward(grad)

    def update(self, lr):
        self.layer1.update(lr)
        self.layer2.update(lr)

    def train(self, X, Y, epochs=1000, lr=0.1):
        for epoch in range(epochs):
            total_loss = 0
            for x, y in zip(X, Y):
                x = x.reshape(-1, 1)
                y = y.reshape(-1, 1)

                loss = self.compute_error(x, y)
                self.backward()
                self.update(lr)

                total_loss += loss

            if epoch % 100 == 0:
                print("Epoch:", epoch, "Loss:", total_loss)


X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

Y = np.array([
    [0],
    [1],
    [1],
    [0]
])

model = SimpleNeuralNetwork()
model.train(X, Y, epochs=2000, lr=0.5)

print("Predictions:")
for x in X:
    output = model.forward(x.reshape(-1, 1))
    print(x, output)
```
